```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('rainbow.jpg',0) # 0 converts it to grayscale
```

```python
plt.imshow(img, cmap='gray')
```

```python
img.max() # in this case the max value is 214
```

```python
ret, thresh1 = cv2.threshold(img,127,255, cv2.THRESH_BINARY) 
#THRESH_BINARY_INV provides inverse
# threshold value = 127, max value = 255
# any value below 127 = 0, any value above 127 = 1
# THRESH_TRUNC above 127 then threshold, else can keep original value
```

```python
ret # displays the cutoff value
```

```python
plt.imshow(thresh1, cmap = 'gray')
```

```python
img = cv2.imread('crossword.jpg',0)
```

```python
plt.imshow(img,cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap = 'gray')
```

```python
show_pic(img)
```

```python
# we want where there is gray to be white
```

```python
ret, th1 = cv2.threshold(img, 150, 255,cv2.THRESH_BINARY)
```

```python
show_pic(th1)
```

```python
# adaptive threshold
```

```python
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY, 11,8)
# blockSize Size of a pixel neighborhood that is used to calculate a threshold value for the
# C Constant subtracted from the mean or weighted mean (see the details below). 
# is positive but may be zero or negative as well.
```

```python
show_pic(th2)
```

```python
blended = cv2.addWeighted(th1,0.6,th2,0.4,0)
show_pic(blended)
```
